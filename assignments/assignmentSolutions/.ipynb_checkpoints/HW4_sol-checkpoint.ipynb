```python
class Rectangle:
    def __init__(self,l,w):
        self.l = l
        self.w = w
    def rectangle_area(self):
        return self.l*self.w
```

```python
R = Rectangle(10,20)
```

```python
R.rectangle_area()
```

```python
%matplotlib notebook
from array import array
import sys
import matplotlib.pyplot as plt
N = []
size_list = []
size_array = []

for num in [2**k for k in range(10, 20)]:
    L = list(range(num))
    size_L = sys.getsizeof(L)

    A = array('i', L)
    size_A = sys.getsizeof(A)

    N.append(num)
    size_list.append(size_L)
    size_array.append(size_A)
    
plt.plot(N, size_array, 'o-')
plt.plot(N, size_list, 's-')
```
