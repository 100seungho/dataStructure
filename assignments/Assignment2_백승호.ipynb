## Problem1
Perform an experimental analysis of the three methods, L.sort(), L.reverse(), len(L), for a list data type. Visualize their running times as a function of the input size. You may reuse the code shown in the lecture.

```python
%matplotlib notebook
from time import time
from random import random
import matplotlib.pyplot as plt
```

```python
n = []
t1 = []
t2 = []
t3 = []
for N in [2 ** k for k in range(15, 25)]:
    L = [random() for i in range(N)]
    
    start_time = time()
    L.sort()
    end_time = time()
    elapsed_time1 = end_time - start_time
    
    start_time = time()
    L.reverse()
    end_time = time()
    elapsed_time2 = end_time - start_time
    
    start_time = time()
    len(L)
    end_time = time()
    elapsed_time3 = end_time - start_time
    
    n.append(N)
    t1.append(elapsed_time1)
    t2.append(elapsed_time2)
    t3.append(elapsed_time3)
    
    print(N, elapsed_time1, elapsed_time2, elapsed_time3)
    
plt.plot(n, t1, 'x-')
plt.plot(n, t2, 'y-')
plt.plot(n, t3, 'o-')
```
