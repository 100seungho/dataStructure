```python
%matplotlib notebook
from time import time
from random import random
import matplotlib.pyplot as plt
```

```python
def prefix_average1(S):
    n = len(S)
    A = [0] * n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += S[i]
        A[j] = total / (j + 1)
    return A
```

```python
def prefix_average3(S):
    n = len(S)
    A = [0] * n
    total = 0
    for j in range(n):
        total += S[j]
        A[j] = total / (j+1)
    return A
```

```python
L = [1,2,3]
```

```python
prefix_average1(L)
```

```python
prefix_average3(L)
```

```python
n = []
t1 = []
t3 = []
for N in [2 ** k for k in range(12, 16)]:
    L = [random() for i in range(N)]
    start_time = time()
    A1 = prefix_average1(L)
    end_time = time()
    elapsed_time1 = end_time - start_time
    start_time = time()
    A3 = prefix_average3(L)
    end_time = time()
    elapsed_time3 = end_time - start_time
    n.append(N)
    t1.append(elapsed_time1)
    t3.append(elapsed_time3)
    print(N, elapsed_time1, elapsed_time3)
    
plt.plot(n, t1, 'x-')
plt.plot(n, t3, 'o-')
```
